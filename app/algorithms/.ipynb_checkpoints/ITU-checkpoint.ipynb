```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from compressive import fourier_cs
import fourier_basis as ft
import matplotlib.pyplot as plt
%matplotlib inline
def plot(data):
    plt.figure()
    sns.tsplot(data)
```

```python
n = 1999
t = np.linspace(0, 1/8, n)
y = np.sin(1394 * np.pi * t) + np.sin(3266 * np.pi * t)
plot(y)
```

```python
freq, rec = fourier_cs(y, 4)
```

```python
fig, ax = plt.subplots(1,2)
sns.tsplot(np.sqrt(1000)*np.abs(ft.half_package(freq)))
sns.tsplot(np.abs(np.fft.fft(y))[0:1000], color='r', alpha = 0.5)
```
